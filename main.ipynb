```python
import numpy as np
import heapq

n = 3

board = np.zeros((n, n))

goal_state = np.zeros((n, n))

for i in range(n):
  for j in range(n):
    goal_state[i, j] = n * i + j + 1

goal_state[n - 1, n - 1] = 0

# n e s w
rowDelta = [-1, 0, 1, 0]
colDelta = [0, 1, 0, -1]

visited = {}

def convertToString(b):
  s = ""
  for i in range(n):
    for j in range(n):
      s += str(b[i, j])

  return s

def hasSolution(b):
  res = 0

  bString = convertToString(b)
  for i in range(n * n):
    for j in range(i + 1, n * n):
      if int(bString[i]) == 0 or int(bString[j]) == 0:
        continue
      if int(bString[j]) < int(bString[i]):
        res += 1

  if res % 2 == 0:
    return True
  return False

for i in range(n):
  s = list(input().split())
  k = 0
  for j in range(len(s)):
    if not str(s[j]).isspace():
      board[i, k] = int(float(s[j]))
      k += 1

board = board.astype(int)

if not hasSolution(board):
  print("Unsolvable :-)")
  raise SystemExit

goal_loc = [0] * (n * n)
for i in range(0, n * n):
  goal_loc[i] = list(zip(*np.where(goal_state == i)))[0]

def getManhattanDist(b):
  curr_loc = [0] * (n * n)
  for i in range(0, n * n):
    #print(b)
    curr_loc[i] = list(zip(*np.where(b == i)))[0]

  mdist = 0

  for i in range(1, (n * n)):
    mdist += abs(curr_loc[i][0] - goal_loc[i][0]) + abs(curr_loc[i][1] - goal_loc[i][1])
  
  return mdist

def getNextBoards(b):
  row = 0
  col = 0

  for i in range(n):
    for j in range(n):
      if b[i, j] == 0:
        row = i
        col = j

  nextB = []
  for d in range(4):
    newRow = row + rowDelta[d]
    newCol = col + colDelta[d]
    if newRow >= 0 and newRow < n and newCol >= 0 and newCol < n:
      newB = np.copy(b)
      newB[row, col] = newB[newRow, newCol]
      newB[newRow, newCol] = 0
      nextB.append(newB)

  return nextB

id = 0

pq = [(getManhattanDist(board), 0, id, board)]

heapq.heapify(pq)

while len(pq) > 0:

  current_state = heapq.heappop(pq)

  if (np.array_equal(current_state[3], goal_state)):
    print(current_state[1])
    break
  
  if (convertToString(current_state[3])) in visited:
    continue

  visited[convertToString(current_state[3])] = True

  for next_board in getNextBoards(current_state[3]):
    id += 1
    priority = getManhattanDist(next_board) + current_state[1] + 1
    moves_so_far = current_state[1] + 1
    heapq.heappush(pq, (priority, moves_so_far, id, next_board))
```
